```python
#hide
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from numpy import pi
import control
rand = np.random.rand
plt.rcParams['font.size'] = 14
```

```python
#hide
def plot_bode(G, f=None):
    if f is None:
        f = np.logspace(-4,3,1000)
    w = 2.0*np.pi*f
    s = 1.0j*w
    Gjw = G(s)
    db = 20.0*np.log10(abs(Gjw))
    phase_rad = np.angle(Gjw)
    phase_rad = np.unwrap(phase_rad)
    phase = phase_rad*180.0/np.pi
    
    plt.figure()
    plt.subplot(211)
    plt.semilogx(f,db)
    plt.ylabel('dB Mag.')
    plt.subplot(212)
    plt.semilogx(f,phase)
    plt.ylabel('Phase (deg.)')
    plt.xlabel('Freq. (Hz)')
```

### Import transfer functions

```python
from bode_gen_examples_2017 import G1ex, G2ex, G3ex
```

## Example 1

```python
plot_bode(G1ex)
```

### Answer

```python
def show_answer(G):
    print('G = %s' % G)
    print('poles (rad/sec) = %s' % np.squeeze(G.pole()))
    print('poles (Hz) = %s' % np.squeeze(G.pole()/(2*pi)))
    print('zeros (rad/sec) = %s' % np.squeeze(G.zero()))
    print('zeros (Hz) = %s' % np.squeeze(G.zero()/(2*pi)))
```

```python
show_answer(G1ex)
```
